```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```
norm = np.random.normal(loc=0, scale=1, size=1000)

fig, ax = plt.subplots(3, 1, sharex=True)

sns.distplot(norm, ax=ax[0])

sns.boxplot(norm, ax=ax[1])

sns.violinplot(norm,ax=ax[2])
plt.show()
```

```
tips_df = pd.read_csv(
    'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
)
tips_df.head()
```

```
sns.violinplot(tips_df['total_bill'])
```

```
sns.violinplot(x='day', y='total_bill', data=tips_df)
```

```
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips_df)
```

```
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips_df, split=True)
```
